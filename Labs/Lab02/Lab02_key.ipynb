## Problem 2

```python
a0_2 = 200000
r_2 = 1.005
n_months_2 = 360
```

For this problem, we have

$$ 200000 = r^0 c + \frac{p}{1-r} $$

and

$$ 0 = r^{360} c + \frac{p}{1-r} $$

This means that

$$ c = 200000 - \frac{p}{1-r} $$

and therefore that 

$$ 0 = 200000 \times r^{360} -  r^{360} \times\frac{p}{1-r} + \frac{p}{1-r} $$

so that 

$$ p = \frac{-200000\times r^{360}(1-r)}{1 - r^{360}} $$

```python
b_2 = -a0_2 * r_2**360. * (1-r_2)/(1 - r_2**360.)
print('Monthly payment should be:', round(b_2,2))
```

So the monthly payment should be about $1,199.10

```python
n_2 = 360
months_2 = np.arange(0, n_2+1)
mortgage_2 = np.empty(n_2+1)
mortgage_2[0] = a0_2
```

```python
for i in range(1, n_2+1):
    mortgage_2[i] = r_2*mortgage_2[i-1] + b_2
    
plt.plot(months_2, mortgage_2, '-');
plt.xlabel(r'Month $n$', fontsize=12);
plt.ylabel(r'Mortage value $a_n$ [dollars]', fontsize=12);
plt.title('Problem 2');
```

```python
print('Amount left at last month = ', round(mortgage_2[n_2],2))
```

So it looks like that solution is correct.

## Problem 4

```python
pop_data = np.array([3.929, 5.308, 7.240, 9.638, 12.866, 
                     17.069, 23.192, 31.443, 38.558, 50.156, 
                     62.948, 75.995,91.972, 105.711, 122.755, 
                     131.669, 150.697, 179.323, 203.212, 
                     226.505, 248.710, 281.416, 308.746])
n_years = len(pop_data)-1
years_data = np.arange(0, n_years+1) # can work with years in units of 10
```

```python
Delta_pn_4 = np.empty(n_years)
pop_data_wout_last = pop_data[0:n_years]
```

```python
for i in range(0,n_years):
    Delta_pn_4[i] = pop_data[i+1] - pop_data[i]
```

```python
plt.plot(pop_data_wout_last, Delta_pn_4, 'o');
plt.xlabel(r'$p_n$ [millions of people]', fontsize=12);
plt.ylabel(r'$\Delta a_n$ [millions of people]', fontsize=12);
plt.title('Problem 4');
```

It doesn't look very linear after the beginning, so try a carrying capacity model instead

```python
P = 330 # capacity of 350 million
pop_data_Pcap = np.empty(n_years)
for i in range(0, n_years):
    pop_data_Pcap[i] = pop_data[i]*(P - pop_data[i])
```

```python
plt.plot(pop_data_Pcap, Delta_pn_4, 'o');
plt.xlabel(r'$p_n(P - p_n)$ [millions of people]${}^2$', fontsize=12);
plt.ylabel(r'$\Delta a_n$ [millions of people]', fontsize=12);
plt.title('Problem 4');
```

```python
Delta_pn_4[0]
```

This looks much better. Let's find a slope and create our population model

```python
k_4 = 0.0009
print('Slope approximately', k_4)
```

```python
Delta_pn_4_pred = np.empty(n_years)
for i in range(0,n_years):
    Delta_pn_4_pred[i] = k_4 * pop_data_Pcap[i]
```

```python
plt.plot(pop_data_Pcap, Delta_pn_4, 'o', label='Population data');
plt.plot(pop_data_Pcap, Delta_pn_4_pred, '-', label='Model prediction')
plt.xlabel(r'$p_n(P - p_n)$ [millions of people]${}^2$', fontsize=12);
plt.ylabel(r'$\Delta a_n$ [millions of people]', fontsize=12);
plt.title('Problem 4 (with prediction)');
```

Return to original data (plot population predictions vs. time)

```python
pn_pred_4 = np.empty(n_years+1)
pn_pred_4[0] = pop_data[0]
for i in range(1,n_years+1):
    pn_pred_4[i] = pn_pred_4[i-1] + k_4 * pn_pred_4[i-1] * (P - pn_pred_4[i-1])
```

```python
plt.plot(years_data, pop_data, 'o', label='Population data');
plt.plot(years_data, pn_pred_4, 'o', label='Model prediction')
plt.xlabel(r'$n$ [tens of years]', fontsize=12);
plt.ylabel(r'$p_n$ [millions of people]', fontsize=12);
plt.title('Problem 4 (with prediction)');
plt.legend(fontsize=12);
```

Prediction looks nearly perfect at beginning but has trouble at the end

```python
perc_err_4 = np.empty(n_years+1)
for i in range(1,n_years+1):
    perc_err_4[i] = abs((pop_data[i] - pn_pred_4[i])/pop_data[i]) * 100.0
```

```python
plt.plot(years_data, perc_err_4, 'o', label='Percent error')
plt.xlabel(r'$n$ [tens of years]', fontsize=12);
plt.ylabel(r'Percent error', fontsize=12);
plt.title('Problem 4 (with prediction)');
```

## Problem 9

Predator-prey model

$$ C_{n+1} = (1+k_1)C_n - k_2 B_n C_n $$
$$ B_{n+1} = (1-k_3)B_n + k_4 B_n C_n $$

### Part (a)

The coefficient $k_1$ represents the growth of the cottony-cushion insect. $k_2$ represents the negative effect that ladybird beetles have on the cottony-cushion insect population because they kill them. $k_3$ represents the decline in the beetle population due to DDT. $k_4$ represents the increase of the beetle population as they eat the cottony-cushion insect.

### Part (b)

The assumption is that the cottony-cushion insect population will naturally increase and that the beetle population will decrease.

### Part (c)

Pick different values of the $k_i$'s 

```python
k1_vals = [0.1, 0.15, 0.17, 0.2]
k2_vals = [0.001, 0.002, 0.002, 0.001]
k3_vals = [0.05, 0.1, 0.06, 0.08]
k4_vals = [0.001, 0.002, 0.001, 0.002]
n_9 = 20

def compute_insect_populations(k1_in, k2_in, k3_in, k4_in, n_in, C0_in, B0_in):
    Cn = np.empty(n_in+1)
    Bn = np.empty(n_in+1)
    Cn[0], Bn[0] = C0_in, B0_in
    for i in range(1, n_in+1):
        Cn[i] = (1 + k1_in)*Cn[i-1] - k2_in * Cn[i-1] * Bn[i-1]
        Bn[i] = (1 - k3_in)*Bn[i-1] + k4_in * Cn[i-1] * Bn[i-1]
    return Cn, Bn
```

```python
Cns_1 = []
Bns_1 = []
C0_1 = 100
B0_1 = 100
for i in range(0,len(k1_vals)):
    Cn_tmp, Bn_tmp = compute_insect_populations(k1_vals[i], k2_vals[i], k3_vals[i], k4_vals[i],\
                                          n_9, C0_1, B0_1)
    Cns_1.append(Cn_tmp)
    Bns_1.append(Bn_tmp)
```

```python
days_9 = np.arange(0,n_9+1)
plt.plot(days_9, Cns_1[0], 'o', label='Cottony-cushion population');
plt.plot(days_9, Bns_1[0], 'o', label='Ladybird beetle population');
plt.xlabel(r'$n$ [days]', fontsize=12);
plt.ylabel(r'Population', fontsize=12);
plt.title('Problem 9, Case 1');
plt.legend(fontsize=12);
```
