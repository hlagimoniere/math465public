## Problem 1

### Part (a)

<img src="images/section2-2_problem6.png" width=200px>

Is $y \propto z^{1/2}$?

```python
y_1a = np.array([3.5, 5, 6, 7, 8])
z_1a = np.array([3,6,9,12,15])

sqrt_z_1a = [np.sqrt(z) for z in z_1a]
```

```python
plt.plot(z_1a, y_1a, 'o');
plt.xlabel(r'$z$', fontsize=12);
plt.ylabel(r'$y$', fontsize=12);
plt.title('y vs. z');
```

```python
plt.plot(sqrt_z_1a, y_1a, 'o');
plt.xlabel(r'$z^{1/2}$', fontsize=12);
plt.ylabel(r'$y$', fontsize=12);
plt.title('y vs. $z^{1/2}$');
```

```python
slope_1a = 2
pred_y_1a = [slope_1a*sqrtz for sqrtz in sqrt_z_1a]

plt.plot(sqrt_z_1a, y_1a, 'o');
plt.plot(sqrt_z_1a, pred_y_1a, '-');
plt.xlabel(r'$z^{1/2}$', fontsize=12);
plt.ylabel(r'$y$', fontsize=12);
plt.title('y vs. $z^{1/2}$');
```

Proportionality works very well for these data

## Problem 2

Working through Example 2 in Section 2.3. Reproduce figure 2.22

<img src="images/section2-3_example2_data.png" width=400px>

Model is $W \propto lg^2$ taking into account girth of the fish in addition to the length.

```python
l_2 = np.array([14.5, 12.5, 17.25, 14.5, 12.625, 17.75, 14.125, 12.625])
g_2 = np.array([9.75, 8.375, 11, 9.75, 8.5, 12.5, 9, 8.5])
W_2 = np.array([27, 17, 41, 26, 17, 49, 23, 16])

n_fish = len(l_2)
lg2_2 = np.empty(n_fish)
for i in range(0, n_fish):
    lg2_2[i] = l_2[i] * g_2[i]**2.0
```

```python
plt.plot(lg2_2, W_2, 'o');
plt.xlabel('$lg^2$ [cubic inches]', fontsize=12);
plt.ylabel('$W$ [ounces]', fontsize=12);
plt.title('$W$ vs. $lg^2$');
```

```python
from scipy import stats
fish_slope, fish_intercept, r_value, p_value, std_err =\
                                stats.linregress(lg2_2, W_2)
print("slope: %f    intercept: %f" % (fish_slope, fish_intercept))
```

```python
W_pred_2 = [fish_slope*lg2 for lg2 in lg2_2]
```

```python
plt.plot(lg2_2, W_2, 'o');
plt.plot(lg2_2, W_pred_2, '-');
plt.xlabel('$lg^2$ [cubic inches]', fontsize=12);
plt.ylabel('$W$ [ounces]', fontsize=12);
plt.title('$W$ vs. $lg^2$');
```

This is the new model with girth included. It seems to capture the structure of the observed data well.
