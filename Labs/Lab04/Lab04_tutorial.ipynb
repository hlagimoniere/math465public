## Using the fitting functions with your data

The file above contains 4 functions. You've used all of them in practice sessions during the lecture.:

- `chebyshev_fit_linear(x_in, data_in, m_in, b_in, xpred_in, display_table=True)` from Lecture23 practice

Here I'll define some test data and show you how to use each of the above functions.

I'm just generating 10 random data points here. For your lab you will use the provided data.

```python
x_test = np.linspace(0,1,num=10)
np.random.seed(25) # so the random numbers are the same every time you run the notebook
y_test = np.random.random(10)
plt.plot(x_test, y_test, 'o');
```

### Define a prediction array

We define an array of more x values so that we can see the straight-line prediction across a wider range of x's than just at the points where data were measured.

Here, since the x data above are generated between 0 and 1, I extend the range to -0.5 to 1.5 and also turn up the number of points.

```python
n_pred = 50
x_pred = np.linspace(-0.5, 1.5, num=n_pred)
```

## Using the visual_fit_linear function

```python
m_guess_1 = 0.1
b_guess_1 = 0.5
visual_fit_linear(x_test, y_test, m_guess_1, b_guess_1, x_pred)
```

## Using the chebyshev_fit_linear function

```python
m_guess_2 = 0.07
b_guess_2 = 0.51
chebyshev_fit_linear(x_test, y_test, m_guess_2, b_guess_2, x_pred, display_table=True)
```

## Using the chebyshev_linprog_fit_linear function

```python
m_guess_3 = 0.07
b_guess_3 = 0.51
chebyshev_linprog_fit_linear(x_test, y_test, x_pred, m_guess_3, b_guess_3)
```

## Using the leastsquares_fit_linear function

```python
m_guess_4 = 0.09
b_guess_4 = 0.3

least_squares_fit_linear(x_test, y_test, m_guess_4, b_guess_4, x_pred, display_table=True)
```
